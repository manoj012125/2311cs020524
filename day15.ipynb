```python
df = sns.load_dataset('taxis')
df
```

```python
sns.barplot(x='payment',y='total',data=df,color='maroon',estimator='mean')
plt.show()
plt.tight_layout()
```

```python
sns.barplot(x='payment',y='total',data=df,color='maroon',estimator='std')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x=df['fare'],color='blue')
```

```python
sns.histplot(x=df['fare'],palette=['blue','green'],hue=df['payment'])
plt.show()
```
